```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions
from selenium.common.exceptions import NoSuchElementException
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
browser.get('https://techstepacademy.com/')
```

```python
def go_to_menu(name_of_menu):
    menu_xpath = '//div[@id="topNav"]//a[text()="'+name_of_menu+'"]'
    menu_element = browser.find_element_by_xpath(menu_xpath)
    ActionChains(browser).click(menu_element).perform()
```

```python
def go_to_menu_about():
    go_to_menu('About')
def go_to_menu_news():
    go_to_menu('News')
def go_to_menu_home():
    go_to_menu('Home')
```

```python
def go_to_blog():
    training_xpath = '//div[@id="topNav"]//a[text()="Training"]'
    live_sessions_xpath = '//div[@id="topNav"]//a[text()="Live Sessions"]'
    menu_element = browser.find_element_by_xpath(training_xpath)
    menu_subelement = browser.find_element_by_xpath(live_sessions_xpath)
    ActionChains(browser).move_to_element(menu_element).click(menu_subelement).perform()
```

```python
def scroll_down():
    footer_xpath = '//footer'
    footer_element = browser.find_element_by_xpath(footer_xpath)
    WebDriverWait(browser, 10).until(expected_conditions.visibility_of(footer_element))
    ActionChains(browser).key_down(Keys.PAGE_DOWN).key_up(Keys.PAGE_DOWN).key_down(Keys.PAGE_DOWN).key_up(Keys.PAGE_DOWN).perform()
```

```python
def make_list_of_blogs():
    go_to_blog()
    all_blogs = browser.find_elements_by_class_name('eventlist-title-link')
    list_of_blogs = []
    for id, blog in enumerate(all_blogs):
        list_of_blogs.append(blog.get_property('href'))
```

```python
def go_to_all_blogs_on_blog_site():
    for blog in list_of_blogs:
        browser.get(blog)
        scroll_down()
```

```python
def go_to_all_blogs_on_blog_site_from_last():
    browser.get(list_of_blogs[-1])
    while True:
        scroll_down()
        last_event_xpath = '//div[@class="eventitem-pager-newer"]//a'
        try:
            last_event_element = browser.find_element_by_xpath(last_event_xpath)
        except NoSuchElementException:
            break
        browser.get(last_event_element.get_property('href'))
    go_to_blog()
```

```python
make_list_of_blogs()
```

```python
go_to_all_blogs_on_blog_site_from_last()
```

```python
go_to_all_blogs_on_blog_site()
```
